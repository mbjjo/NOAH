When a simulation is run without interactions like above no errors are shown in Python. Check the report file to see if the simulation is run correctly. 

Below is shown an example of how data can be stored during the simulation.

This script computes the water level (depth) and flooding in node N4 and the flow through link 2. 

The model used is case1. 

```python
with Simulation("./case1.inp",'case1.rpt','case1.out') as sim: 
    # The elements, such as nodes and links, that are used in the simulation are defined. The orifice (reg1) is defined because it is used in the control algorithm.
    # The elements are defined accordingly to their names in the SWMM model
    # This is not strictly necessary but makes the later part of the code easier to read. 
    system = SystemStats(sim)
    N4 = Nodes(sim)["N4"]
    link2 = Links(sim)["link2"]
    reg1 = Links(sim)["reg1"]
    # Empty lists are created to contain data from each simulation step.
    # Empty lists for the time data are also created. 
    time_hr = list()
    time_sec = list()
    N4_depth = list()
    N4_flood = list()
    link2_flow = list()

    # The simulation is initialized. 
    # The following code is repeated for every step i the simulation
    for step in sim:
        # Control algorithm is computed. 1 is a constant in this example,
        # but could also be a variable that can be modified.
        if N4.depth >= 1:
            reg1.target_setting = 0
        else:
            reg1.target_setting = 1

        # Time is computed to be used in plots. Both time in hours and seconds are stored,
        # since they are used for different purposes.
        dt = sim.current_time
        time_hr.append(dt.hour + dt.minute/60 + dt.second/60/60)
        time_sec.append(dt.hour*60*60 + dt.minute*60 + dt.second) 
        # Depth, flow and other data is computed and stored in the lists created earlier. 
        N4_depth.append(N4.depth)
        N4_flood.append(N4.flooding)
        link2_flow.append(link2.flow)
    # The simulation ends

    # Data statistics (routing and runoff) are stored in dictionaries. 
    # This contains the statistics that are also available in the report file from SWMM
    routing = system.routing_stats
    # Specific statistics can be accessed like shown below. 
    total_flooding = routing['flooding']
    runoff = system.runoff_stats
    N4_floodStat = N4.statistics['flooding_volume']
    # Some data such as flooding needs numerical integration to provide a result of the total amount of flooding throughout the simulation.
    # Time in seconds is used here because the unit of e.g. flooding is m3/s
    time = time_sec
    N4_flooding = integrate.simps(N4_flood,x = time)
```

The computed data is shown in plots below.

The time in hours is used on the x-axis. Notice that when doing simulations for longer than one day time needs to be computed differently.

```python
fig = plt.figure()
plt.ylabel('Water level in node N4 [m]',fontsize = 14)
plt.xlabel('Time of simulation [hours]',fontsize = 14)
plt.plot(time_hr,N4_depth,linewidth =2)
plt.yticks(fontsize = 14)
plt.xticks(fontsize = 14)
plt.legend(('Water level in node N4 [m]',),fontsize = 10)
plt.tight_layout()
# fig.savefig('figure_name.png')
plt.show()
```

```python
fig = plt.figure()
plt.ylabel('flow in link 2 [m3/s]',fontsize = 14)
plt.xlabel('Time of simulation [hours]',fontsize = 14)
plt.plot(time_hr,link2_flow,linewidth =2)
plt.yticks(fontsize = 14)
plt.xticks(fontsize = 14)
plt.legend(('flow in link 2 [m3/s]',),fontsize = 10)
plt.tight_layout()
# fig.savefig('figure_name.png')
plt.show()
```

#### Running simulations as definitions 
Simulations can be defined as functions to make computing easier to read and write.

The code is similar to the one above, with the exception of variables that can be assigned to each simulation. 
In the example is the variable assigned to the control algorithm but other parameters can also be computed with the variable. Also several variables can be used for each simulation.

```python
def simulation(x):
    with Simulation("./case1.inp",'case1.rpt','case1.out') as sim:
        # ...
        # Similar code as above, but the constant is changed to a variable in the control algorithm
        if N4.depth > x:
            reg1.target_setting = 0
        else:
            reg1.target_setting = 1
        # ...
    # The desired output parameters are specified
    return time_hr, N4_depth, link2_flow, N4_flooding, routing, runoff
```

#### Error messages 
If error massages occur check the generated .rpt-file or messages in the output.

For a full list of error messages see appendix E in the User's Manual on the SWMM website. 

Make sure that the specified path in the SWMM model is correct when using other files. This applies to files such as time series with rain data and backdrop files with maps. 

#### Other comments

One advantage with SWMM is that the .inp-files can be modified in a normal text editor making it easier to apply changes to the model.  

Many functions are available in the pyswmm package such as average, maximum and total values. Check the documentation to see if there is anything that will fit your purpose before calculating it yourself. 

Note that the total flooding in the report file is the amount of water that is lost from the system during flooding. If water is allowed to reenter the system (if "allow ponding" is checked) it will not count in this statistics even though flooding occur from these nodes. 

Control rules can be computed in the Control Rule Editor in SWMM. Make sure that this is empty if RTC is applied through pyswmm. Pyswmm does not owerwrite control rules computed in SWMM. 

Use the User Manuals available on the SWMM website. Here you will find list of error codes, describtions of the used equations and methods and examples on how to use the SWMM interface. Also examples of SWMM models are available in the SWMM Aplication Manual. (See https://www.epa.gov/water-research/storm-water-management-model-swmm) 

A new version of pyswmm came out recently. Make sure to have the updated version (0.5.0).
